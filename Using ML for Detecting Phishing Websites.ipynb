# Creating a Logistic Regression Model

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
from sklearn.linear_model import LogisticRegression
import wandb 
import time 
```

```python
# we create a utility function that we can reuse
```

```python
def train_eval_pipeline(model, train_data, test_data, name):
    #initialize wandb
    wandb.init(project = 'Machine Learning for phishing website detection', name=name)
    #assign the data
    (X_train, y_train)=train_data
    (X_test, y_test)=test_data

    #Train the model
    start=time.time()
    model.fit(X_train, y_train)
    end=time.time()-start
    prediction=model.predict(X_test)

    wandb.log({"accuracy":accuracy_score(y_test, prediction)*100, "precision":precision_recall_fscore_support(y_test, prediction, average='macro')[0], "recall":precision_recall_fscore_support(y_test, prediction, average='macro')[1], "training time": end})
    print("Accuracy Score of the Logistic Regression Classifier with default hyperparameter values {0:.2f}%".format(accuracy_score(y_test, prediction)*100.))
    print("\n")
    print("---Classification report of the Logistic Regression Classifier with default parameter values---")
    print("\n")
    print(classification_report(y_test, prediction, target_names=["Phishing Websites", "Normal Websites"]))
```

```python
logreg = LogisticRegression()
```

```python
logreg
```

```python
train_eval_pipeline(logreg, (X_train, y_train), (X_test, y_test), "Logistic_Regression_Phishing_Websites")
```

# Can we improve the model?

A good way to start approaching this is to tune the hyperparameters of the model. We need to define the grid of the values for the hyperparameters that we need to tune. For this, we use random search for hyperparameter tuning

```python
# Import the GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#define the grid of values
penalty = ["l1", "l2"]
C = [0.8, 0.9, 1.0]
tol = [0.01, 0.001, 0.0001]
max_iter = [100, 150, 200, 250]
```

```python
#Create a dict where tol and max_iter are keys and lists of their values are the corresponding values
param_grid = dict(penalty = penalty, C=C, tol = tol, max_iter = max_iter)
```

```python
#Now that we have the grid, we look for a set of hyperparameter values
#We instantiate RandomizedCV with the required paramters
random_model = RandomizedSearchCV(estimator=logreg, param_distributions=param_grid, cv=5)
```

```python
#Fit the model to the data
random_model_result = random_model.fit(X_train,y_train)
```

```python
#summarize the results
best_score, best_params = random_model_result.best_score_,random_model_result.best_params_
```

```python
print('Best Score: %.2f using %s'% (best_score*100.,best_params))
```

Comments - Random Search did not help much in boosting accuracy

```python
#Log the results of the hyperparameter in wandb
config = wandb.config
config.tol = 0.01
config.penalty = "l2"
config.C = 0.9
```

```python
#Train the model
logreg = LogisticRegression(tol = config.tol, penalty=config.penalty, max_iter= 100, C=config.C)
```

```python
train_eval_pipeline(logreg, (X_train, y_train), (X_test, y_test), "Logistic_Regression_Random_Search")
```
